```python
y_train = df_train['time'].values
```

```python
X_train = df_train.drop(['place', 'time'], axis=1).values
```

```python
y_test = df_test['time'].values
X_test = df_test.drop(['place', 'time'], axis=1).values
```

```python
X_train
```

```python
import xgboost as xgb
```

```python
train = xgb.DMatrix(X_train, label=y_train)
test = xgb.DMatrix(X_test, label=y_test)
```

```python
xgb_params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse'}
```

```python
bst = xgb.train(
        xgb_params,
        train,
        num_boost_round=100)
```

```python
y_pred = bst.predict(test)
```
